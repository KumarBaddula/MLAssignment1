```python
import statistics

ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24] 

print("-----Question 1-----\n")

''' Sort the list and find the min and max age  '''

ages.sort()                 #sort() is the method used in python to sort a list

print('Sorted list: ', ages, '\n')                 #print("") is the method to show output on the screen

print('Minimum Age: ', min(ages), '\n')            #min() is the method to find the smallest number in a list

print('Maximum Age: ', max(ages), '\n')            #max() is the method to find the largest number in a list


''' Add min age and max age again to the list '''

ages.append(min(ages))      #append() is the method used to add a value to the list
ages.append(max(ages))

print('List after Adding Elements: ', ages, '\n')


''' Find the median age (one middle item or two middle items divided by two) '''

print('Median age: ', statistics.median(ages), '\n')     #statistics is a package which consists of methods to find mean median etc..,


''' Find the average age (sum of all items divided by their number) '''

print('Average: ', sum(ages)/len(ages), '\n')          #sum() method is used to add all the values inside a list
                                    		       #len() methos is used to find number of elements present inside a list


''' Find the range of the ages (max minus min) '''

print('Range: ', max(ages) - min(ages), '\n')
```

```python
''' ----- Question 5----- '''
print("----- Question 5 -----", '\n')

import math

radius = 30

''' Calculate the area of a circle and assign the value to a variable name of _area_of_circle_ '''

_area_of_circle_ = math.pi*radius*radius
print('Area of the Circle: ', _area_of_circle_, '\n')


''' Calculate the circumference of a circle and assign the value to a variable name of_circum_of_circle_ '''

_circum_of_circle_ = 2*math.pi*radius
print('Circumference of the Circle: ', _circum_of_circle_, '\n')


''' Take radius as user input and calculate the area. '''

print('Enter Radius: ')
radius = float(input())
print('Area of the Circle: ', math.pi*radius*radius)

```

```python
''' ----- Question 9----- '''

print("----- Question 9 -----", '\n')

'''Write a program, which reads weights (lbs.) of N students into a list and convert these weights to 
kilograms in a separate list using Loop. N: No of students (Read input from user)
Ex: L1: [150, 155, 145, 148]
Output: [68.03, 70.3, 65.77, 67.13] '''

print("No. of Students: ")
N = int(input())
std_weights_lbs = []
for i in range(0, N):
    A = float(input())
    std_weights_lbs.append(A)
print('List of Weights of Students in lbs: ', std_weights_lbs, '\n')
std_weights_kg = []
for i in range(0, N):
    B = std_weights_lbs[i]*0.454
    std_weights_kg.append(B)
print('List of Weights of Students in kgs: ', std_weights_kg, '\n')
```
